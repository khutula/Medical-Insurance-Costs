```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_path = "../../insurance.csv"
df = pd.read_csv(csv_path)
print(df.dtypes)
print(df.nunique())
df
```

```python
plt.bar(df['bmi'], df['charges'])
plt.title('BMI vs Charges')
plt.show()
```

```python
plt.bar(df['age'], df['charges'])
plt.title('Age vs Charges')
plt.show()
```

```python
df.describe()
```

```python
df = df.reset_index()
df = pd.get_dummies(df)
df
```

```python
plt.bar(df['index'], df['charges'])
plt.title('Index vs Charges')
plt.show()
```

```python
pd.qcut(df['charges'], q=10)
```

```python
df['charges_quantile'] = pd.qcut(df['charges'], q=10)
df
```

```python
groupby_count_df = df.groupby('charges_quantile').count()
groupby_count_df
```

```python
groupby_mean_df = df.groupby('charges_quantile').mean()
groupby_mean_df
```

```python
plt.bar(groupby_mean_df['bmi'], groupby_mean_df['charges'])
plt.title('BMI vs Charges')
plt.show()
```
